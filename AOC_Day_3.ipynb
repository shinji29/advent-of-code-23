### Part 1 _Solution_

```python
s = [list(x) for x in data]
compact(s)
```

#### How to compute the _ADJACENCY_ in the matrix ...

```python
flag = False

adj = [(dx, dy) for dx in (-1, 0, 1) for dy in (-1, 0, 1) if (dx, dy) != (0, 0)]

example = [
    ['1', '.', '.'],
    ['.', '.', '.'],
    ['.', '.', '@']
]

for i, row in enumerate(example):
    for j in range(len(example[0])):
        x = row[j]
        if x.isdigit():
            if any(
                example[i+x][j+y] != '.' and not example[i+x][j+y].isdigit()
                for (x, y) in adj
                if i+x in range(len(example)) and j+y in range(len(example[0]))
            ):
                flag = True
                break

print("Adjacent engine component exists!") if flag else print("No adjacent engine component exists!")
```

#### Coding the solution ...

```python
ans1 = 0
thing = ""
keep = False

for i, r in enumerate(s):
    
    for j in range(len(r)):
        v = r[j]
        
        if v.isdigit():
            
            if any(s[i+dx][j+dy] != "." and not s[i+dx][j+dy].isdigit()
                for dx, dy in adj 
                if (i+dx) in range(len(s)) and (j+dy) in range(len(s[i+dx]))):
                keep = True
            thing += v
            
        else:
            if keep and len(thing) > 0:
                ans1 += int(thing)
                
            thing = ""
            keep = False

if keep and len(thing) > 0:
    ans1 += int(keep)
```

### Part 2 _Solution_

```python
class D:
    def __init__(self, a):
        self.a = ""
        
s2 = [[None for _ in r] for r in s]
thing = D("")

for i, r in enumerate(s):
    
    for j in range(len(r)):
        v = r[j]
        
        if v.isdigit():
            thing.a += v
            s2[i][j] = thing.a
            
        else:
            thing = D("")
```

#### s2 represents the matrix containing only those digits that are adjacent to symbols

```python
compact(s2)
```

```python
ans2 = 0

for i, r in enumerate(s):
    for j, v in enumerate(r):
        if v == "*":
            adj_nums = []
            
            for x, y in adj:
                if i+x in range(len(s)) and j+y in range(len(s[i+x])):
                    
                    if s2[i+x][j+y] is not None:
                        vv = s2[i+x][j+y]
                    
                        if vv not in adj_nums:
                            adj_nums.append(vv)
                        
            if len(adj_nums) == 2:
                ans2 += int(adj_nums[0].a) * int(adj_nums[1].a)
```
